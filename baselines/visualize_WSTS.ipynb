```python
import rasterio
import matplotlib.pyplot as plt

tiff_path = "/Users/natashayang/FireNet/data/raw/fire_24935867/2021-02-28.tif"

# Open and read the TIFF file
with rasterio.open(tiff_path) as src:
    img = src.read(1)  # Read the first band (grayscale)
    print(f"Width: {src.width}, Height: {src.height}")
    print(f"Number of Bands: {src.count}")
    print(f"Data Type: {src.dtypes[0]}")
    print(f"CRS (Coordinate Reference System): {src.crs}")

# Display the image
plt.figure(figsize=(8, 8))
plt.imshow(img, cmap="gray")
plt.colorbar()
plt.title("TIFF Visualization")
plt.axis("off")
plt.show()

```
